##機械学習の前処理

```python
#変数の再確認
df_train_master.info()
```

```python
df_train_master.head(10)
```

```python
#機械学習用のデータフレームを作成
df_train = df_train_master.copy()
df_submit = df_submit_master.copy()

#carIDは削除
df_train = df_train.drop("carID", axis=1)

#brand 一番なじみのあるtoyotaを除く8ブランドをダミー化(one-hot encoding)
df_train["brand_vauxhall"] = df_train["brand"].apply(lambda x: 1 if x == "vauxhall" else 0)
df_train["brand_merc"] = df_train["brand"].apply(lambda x: 1 if x == "merc" else 0)
df_train["brand_hyundi"] = df_train["brand"].apply(lambda x: 1 if x == "hyundi" else 0)
df_train["brand_vw"] = df_train["brand"].apply(lambda x: 1 if x == "vw" else 0)
df_train["brand_bmw"] = df_train["brand"].apply(lambda x: 1 if x == "bmw" else 0)
df_train["brand_ford"] = df_train["brand"].apply(lambda x: 1 if x == "ford" else 0)
df_train["brand_skoda"] = df_train["brand"].apply(lambda x: 1 if x == "skoda" else 0)
df_train["brand_audi"] = df_train["brand"].apply(lambda x: 1 if x == "audi" else 0)
df_train = df_train.drop("brand", axis=1)
df_submit["brand_vauxhall"] = df_submit["brand"].apply(lambda x: 1 if x == "vauxhall" else 0)
df_submit["brand_merc"] = df_submit["brand"].apply(lambda x: 1 if x == "merc" else 0)
df_submit["brand_hyundi"] = df_submit["brand"].apply(lambda x: 1 if x == "hyundi" else 0)
df_submit["brand_vw"] = df_submit["brand"].apply(lambda x: 1 if x == "vw" else 0)
df_submit["brand_bmw"] = df_submit["brand"].apply(lambda x: 1 if x == "bmw" else 0)
df_submit["brand_ford"] = df_submit["brand"].apply(lambda x: 1 if x == "ford" else 0)
df_submit["brand_skoda"] = df_submit["brand"].apply(lambda x: 1 if x == "skoda" else 0)
df_submit["brand_audi"] = df_submit["brand"].apply(lambda x: 1 if x == "audi" else 0)
df_submit = df_submit.drop("brand", axis=1)

#model バリデーション用のtrainデータとtestデータに分割した後に処理するため、今は処理なし

#year　1970年のデータだけ削除
df_train = df_train.query('year != 1970')

#transmission 一番なじみのあるAutomaticを除く3カテゴリーをダミー化(one-hot encoding)
df_train["transmission_manual"] = df_train["transmission"].apply(lambda x: 1 if x == "Manual" else 0)
df_train["transmission_semi-auto"] = df_train["transmission"].apply(lambda x: 1 if x == "Semi-Auto" else 0)
df_train["transmission_other"] = df_train["transmission"].apply(lambda x: 1 if x == "Other" else 0)
df_train = df_train.drop("transmission", axis=1)
df_submit["transmission_manual"] = df_submit["transmission"].apply(lambda x: 1 if x == "Manual" else 0)
df_submit["transmission_semi-auto"] = df_submit["transmission"].apply(lambda x: 1 if x == "Semi-Auto" else 0)
df_submit["transmission_other"] = df_submit["transmission"].apply(lambda x: 1 if x == "Other" else 0)
df_submit = df_submit.drop("transmission", axis=1)

#mileage 逆数変換
df_train["mileage"] = 1 / (df_train["mileage"] + 10000)
df_submit["mileage"] = 1 / (df_submit["mileage"] + 10000)

#fuelType Dieselを除く4カテゴリーをダミー化
df_train["fueltype_petrol"] = df_train["fuelType"].apply(lambda x: 1 if x == "Petrol" else 0)
df_train["fueltype_petrol"] = df_train["fuelType"].apply(lambda x: 1 if x == "Hybrid" else 0)
df_train["fueltype_petrol"] = df_train["fuelType"].apply(lambda x: 1 if x == "Other" else 0)
df_train["fueltype_petrol"] = df_train["fuelType"].apply(lambda x: 1 if x == "Electric" else 0)
df_train = df_train.drop("fuelType", axis=1)
df_submit["fueltype_petrol"] = df_submit["fuelType"].apply(lambda x: 1 if x == "Petrol" else 0)
df_submit["fueltype_petrol"] = df_submit["fuelType"].apply(lambda x: 1 if x == "Hybrid" else 0)
df_submit["fueltype_petrol"] = df_submit["fuelType"].apply(lambda x: 1 if x == "Other" else 0)
df_submit["fueltype_petrol"] = df_submit["fuelType"].apply(lambda x: 1 if x == "Electric" else 0)
df_submit = df_submit.drop("fuelType", axis=1)

#tax 100<tax<200をベースに「tax<=100,200<=tax」の2カテゴリーをダミー化
df_train["tax_100under"] = df_train["tax"].apply(lambda x: 1 if x <= 100 else 0)
df_train["tax_200over"] = df_train["tax"].apply(lambda x: 1 if x >= 200 else 0)
df_train = df_train.drop("tax", axis=1)
df_submit["tax_100under"] = df_submit["tax"].apply(lambda x: 1 if x <= 100 else 0)
df_submit["tax_200over"] = df_submit["tax"].apply(lambda x: 1 if x >= 200 else 0)
df_submit = df_submit.drop("tax", axis=1)

#mpg 対数変換したmpgで3.5~4をベースに「"3.6以下","4以上"」の2カテゴリーをダミー化
df_train["log_mpg_3.6under"] = df_train["mpg"].apply(lambda x: 1 if np.log(x) <= 3.6 else 0)
df_train["log_mpg_4.0over"] = df_train["mpg"].apply(lambda x: 1 if np.log(x) >= 4 else 0)
df_train = df_train.drop("mpg", axis=1)
df_submit["log_mpg_3.6under"] = df_submit["mpg"].apply(lambda x: 1 if np.log(x) <= 3.6 else 0)
df_submit["log_mpg_4.0over"] = df_submit["mpg"].apply(lambda x: 1 if np.log(x) >= 4 else 0)
df_submit = df_submit.drop("mpg", axis=1)

#engineSize 処理なし

#データの確認
display(df_train.info())
display(df_train.describe())
display(df_submit.info())
display(df_submit.describe())
```

```python
fig = plt.subplots(figsize=(20,10))
sns.heatmap(df_train.corr(), cmap="bwr", annot=True, vmax=1, vmin=-1, center=0)
#多重共線性は問題なさそう。単変量比較ではpriceと一番相関が強いのはengineSize
```

```python
#説明変数と目的変数に分割
X = df_train.drop("price", axis=1)
y = df_train["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

#X_trainとy_trainを使って各modelのpriceの中央値を算出、model_scoreとしてmergeする
model_df = pd.merge(X_train, y_train, left_index=True, right_index=True)[["model", "price"]]
model_df = model_df.groupby("model", as_index=False).median().rename(columns={"price": "model_score"})
X_train = pd.merge(X_train, model_df, on="model", how="left")
X_train = X_train.drop("model", axis=1)
X_test = pd.merge(X_test, model_df, on="model", how="left")
X_test = X_test.drop("model", axis=1)
display(X_train.info())
display(X_test.info())
```

##予測と評価

```python
#重回帰(sklearn)
from sklearn.linear_model import LinearRegression
lr = LinearRegression() 
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

#スコア算出
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("r2=", r2_score(y_true=y_test, y_pred=y_pred))
print("RMSE=", np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_pred)))
print("MAE=", mean_absolute_error(y_true=y_test, y_pred=y_pred))
pd.DataFrame(lr.coef_, index=X_train.columns)
```

```python
#標準化偏回帰係数を確認

#標準化
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_std = ss.fit_transform(X_train)
X_train_std = pd.DataFrame(X_train_std,columns=X_train.columns)
X_test_std = ss.transform(X_test)
X_test_std = pd.DataFrame(X_test_std,columns=X_test.columns)
lr2 = LinearRegression() 
lr2.fit(X_train_std, y_train)
y_pred = lr2.predict(X_test_std)

#標準化偏回帰係数を算出
pd.DataFrame(lr2.coef_, index=X_train_std.columns)

#model, mileage, yearがscoreの予測に寄与
```

```python
#ベースラインモデル math scoreの中央値を代入して予測
y_pred = [np.median(y_train)] * len(y_test)
print("RMSE=", np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_pred)))
#重回帰モデルのスコアがベースラインモデルを上回っていることを確認
```

```python
#決定木分析
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=20) #max_depthを指定
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

#スコア算出
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("r2=", r2_score(y_true=y_test, y_pred=y_pred))
print("RMSE=", np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_pred)))
print("MAE=", mean_absolute_error(y_true=y_test, y_pred=y_pred))
pd.DataFrame(lr.coef_, index=X.columns)

#RMSEの低い決定木を採用
```

```python
#誤差の多かったデータと誤差の少なかったデータで比較を行う
X_test["y_pred"] = y_pred
X_test["y_true"] = np.array(y_test)
X_test["error"] = X_test["y_pred"] - X_test["y_true"]
X_test["abs_error"] = abs(X_test["y_pred"] - X_test["y_true"])
display(X_test.sort_values("abs_error", ascending=False).head(10))
display(X_test.sort_values("abs_error", ascending=True).head(10))
```

```python
#誤差の分布を確認
fig = plt.subplots(figsize=(20,10))
sns.distplot(X_test["abs_error"], kde=False)

#基本的には誤差が少ない場合が多いが、時々誤差が大きくなってしまうケースが存在する様子
```

```python
#誤差とpriceの関係性を確認

fig = plt.subplots(figsize=(20,10))
sns.scatterplot(x="y_true", y="error", data=X_test)
plt.axhline(y=0,c="gray",linestyle="--")

#高価な車ほど誤差が大きい傾向あり、高価な車の誤差を減らすのが必要になるか
```

```python
#submit用モデルを作るためにtrainデータ全てを使用して再学習を実行
model_df = df_train_master[["model", "price"]]
model_df = model_df.groupby("model", as_index=False).median().rename(columns={"price": "model_score"})
X = pd.merge(X, model_df, on="model", how="left")
X = X.drop("model", axis=1)
df_submit = pd.merge(df_submit, model_df, on="model", how="left")
df_submit = df_submit.drop("model", axis=1)
clf = tree.DecisionTreeClassifier(max_depth=20) #max_depthを指定
clf = clf.fit(X,y)
```

```python
#作成したモデルでsubmitデータを予測
X_submit = df_submit.drop("carID", axis=1)
pred = clf.predict(X_submit)
df_submit["y"] = pred
submit_data = df_submit[["carID", "y"]].set_index("carID")
submit_data.to_csv("car_price_prediction_submit.csv", index=True)
submit_data
#最終スコア6492
```
