## 2.0-Attrition

## 2.1- Age

```python
# 　Age
fig = plt.subplots(figsize=(20,10))
sns.histplot(train.query("Attrition == 1")["Age"],kde=False,label="Attrition",color="red")
sns.histplot(train.query("Attrition == 0")["Age"],kde=False,label="Attrition",color="blue")


plt.legend();
#train.groupby(["Attrition"]).mean()["Age"]
train.groupby("Attrition")["Age"].mean()
```

## 2.2- BusinessTrabel

```python
# BusinessTrabel
fig = plt.subplots(figsize=(20,10))
sns.countplot(x="BusinessTravel",hue="Attrition",data=train)
#train.groupby(["BusinessTravel"]).mean()["Attrition"]
train.groupby("BusinessTravel")["Attrition"].mean()


# 出張頻度が高いほど退職率が高くなる
# Non=1,Rere=2,Frec=3の順序尺度に変換が良さそう
```

## 2.3-MonthlyIncome

```python
train.query('MonthlyIncome == 0')

# 中央値で埋めるという手段もあるが、1行だけなので削除で良さそう
```

```python
# MonthlyIncome 
fig = plt.subplots(figsize=(20,10))
sns.histplot(train.query("Attrition == 1")["MonthlyIncome"],kde=False,label="Attrition",color="red")
sns.histplot(train.query("Attrition == 0")["MonthlyIncome"],kde=False,label="Attrition",color="blue")
plt.legend()
train.query("MonthlyIncome != 0").groupby(["Attrition"])["MonthlyIncome"].mean()
```

## 2.4- Department

```python
# Department
fig = plt.subplots(figsize=(20,10))
sns.countplot(x="Department",hue="Attrition",data=train)
train.groupby(["Department"])["Attrition"].mean()
```

## 2.5- DistanceFromHome

```python
# DistanceFromHome
fig = plt.subplots(figsize=(20,10))
sns.histplot(train.query("Attrition == 1")["DistanceFromHome"],kde=False,label="Attrition",color="red",bins=30)
sns.histplot(train.query("Attrition == 0")["DistanceFromHome"],kde=False,label="Attrition",color="blue",bins=30)
plt.vlines(21,0,150,"gray",linestyle="--")
plt.legend()
train.groupby(["Attrition"])["DistanceFromHome"].mean()


# 職場と家までの距離が21(マイル?)を超えると退職者の割合が一気に増える
# この特徴量は残しつつ、21マイル以上（遠距離）かどうか判定するダミー変数の追加を検討
```

## 2.6- EdudcationField

```python
# EdudcationField

fig = plt.subplots(figsize=(20,10))
sns.countplot(x="EducationField",hue="Attrition",data=train)
train.groupby(["EducationField"])["Attrition"].mean()
```

## 2.7-EmployeeCount

```python
# EmployeeCount
train["EmployeeCount"].describe
# 全て1なので意味なし。特徴量から削除
```

```python
# EnvironmentSatisfactionの調査
display(train.query("EnvironmentSatisfaction != EnvironmentSatisfaction"))
train.query("EnvironmentSatisfaction != EnvironmentSatisfaction")["Attrition"].mean()

# 欠損しているデータに共通していそうな項目はないが、退職率は18%と欠損を含まないデータより高い印象
# このデータがアンケートで収集されたものだとしたら、そもそも退職以降の強い人はアンケートに協力していない可能性がある
```

```python
# EnvironmentSatisfaction

fig = plt.subplots(figsize=(20,10))
sns.histplot(train.query("Attrition == 1")["EnvironmentSatisfaction"],kde=False,label="Attrition",color="red",bins=4)
sns.histplot(train.query("Attrition == 0")["EnvironmentSatisfaction"],kde=False,label="Attrition",color="blue",bins=4)
plt.legend()
train.groupby(["EnvironmentSatisfaction"])["Attrition"].mean()

# 満足度が1点だと退職率が高くなる傾向あり
# 満足度が1点か否か、満足度が欠損か否かのダミー変数を作成する方針
```

## 2.11-NumCompaniesWorked	

```python
# NumCompaniesWorked(経験している会社数)
fig = plt.subplots(figsize=(20,10))
sns.countplot(x="NumCompaniesWorked",hue="Attrition",data=train)
plt.legend()
train.groupby(["NumCompaniesWorked"])["Attrition"].mean()

# 経験している会社数が多いほど退職率が多い
```
