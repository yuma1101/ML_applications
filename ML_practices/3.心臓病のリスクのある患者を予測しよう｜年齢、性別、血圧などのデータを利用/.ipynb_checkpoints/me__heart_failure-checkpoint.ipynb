# 2.データ探索(EDA)

```python
def quick_eda(df):
    print("====================================")
    print("① 基本情報（info）")
    print("====================================")
    print(df.info())

    print("\n====================================")
    print("② 行数・列数（shape）")
    print("====================================")
    print(df.shape)

    print("\n====================================")
    print("③ 基本統計量（describe）")
    print("====================================")
    print(df.describe(include="all"))

    print("\n====================================")
    print("④ 欠損値の合計")
    print("====================================")
    print(df.isnull().sum())

    print("\n====================================")
    print("⑤ 欠損値の割合（%）")
    print("====================================")
    print((df.isnull().mean() * 100).round(2))

    print("\n====================================")
    print("⑥ 重複行の数")
    print("====================================")
    print(df.duplicated().sum())

    print("\n====================================")
    print("⑦ カテゴリ列のユニーク値数")
    print("====================================")
    cat_cols = df.select_dtypes(include=["object", "category"]).columns
    for col in cat_cols:
        print(f"{col}: {df[col].nunique()} unique values")

    print("\n====================================")
    print("⑧ 列名一覧")
    print("====================================") 
    print(df.columns.tolist())

    print("\n====================================")
    print("⑨ 数値列の相関係数")
    print("====================================")
    print(df.corr(numeric_only=True))

    print("\n====================================")
    print("⑩ メモリ使用量（MB）")
    print("====================================")
    print(df.memory_usage(deep=True).sum() / 1024**2, "MB")

    print("\n====================================")
    print("⑪ データの先頭と末尾の5行（head）")
    print("====================================")
    print(df.head())
    print(df.tail())

    print("\n====================================")
    print("⑫ 欠損値ヒートマップ（可視化）")
    print("====================================")
    plt.figure(figsize=(10, 6))
    sns.heatmap(df.isnull(), cbar=False)
    plt.title("Missing Value Heatmap")
    plt.show()

    print("\n====================================")
    print("⑬ 数値列ヒストグラム（可視化）")
    print("====================================")
    df.hist(figsize=(12, 8), bins=20)
    plt.tight_layout()
    plt.show()

    print("\n====================================")
    print("⑭ ランダムサンプル ")
    print("====================================")    
    display(df.sample(head))
```

```python
quick_eda(train)
```

# 3.機械学習の前処理

```python
#変数の再確認
train.info()
```

```python
df_train = train.copy()
df_test = test.copy()

#ID
df_train = df_train.drop('ID',axis=1)

#Sex 男性を1、女性を0に
df_train['Sex'] = df_train['Sex'].apply(lambda x:1 if x == 'M' else 0)
df_test['Sex'] = df_test['Sex'].apply(lambda x:1 if x == 'M' else 0)

#ChestPainType それぞれのカテゴリーでダミー変数を作成する
df_train['ChestPainType_ASY'] = df_train['ChestPainType'].apply(lambda x:1 if x =='ASY' else 0)
df_train['ChestPainType_NAP'] = df_train['ChestPainType'].apply(lambda x:1 if x =='NAP' else 0)
df_train['ChestPainType_ATA'] = df_train['ChestPainType'].apply(lambda x:1 if x =='ATA' else 0)
df_train = df_train.drop('ChestPainType',axis=1)
df_test['ChestPainType_ASY'] = df_test['ChestPainType'].apply(lambda x:1 if x =='ASY' else 0)
df_test['ChestPainType_NAP'] = df_test['ChestPainType'].apply(lambda x:1 if x =='NAP' else 0)
df_test['ChestPainType_ATA'] = df_test['ChestPainType'].apply(lambda x:1 if x =='ATA' else 0)
df_test = df_test.drop('ChestPainType',axis=1)

#RestingBP nullのデータを削除
df_train = df_train.dropna(subset = ['RestingBP'])

#Cholesterol trainデータの平均値で補完(df_submitもtrainデータの平均値で補完)
df_train['Cholesterol'] = df_train['Cholesterol'].fillna(np.mean(df_train['Cholesterol']))
df_test['Cholesterol'] = df_test['Cholesterol'].fillna(np.mean(df_train['Cholesterol']))

#RestingECG　ダミー変数化
df_train['RestingECG_ST'] = df_train['RestingECG'].apply(lambda x:1 if x== 'ST' else 0)
df_train = df_train.drop('RestingECG',axis=1)
df_test['RestingECG_ST'] = df_test['RestingECG'].apply(lambda x:1 if x== 'ST' else 0)
df_test = df_test.drop('RestingECG',axis=1)

#ExerciseAngina ダミー変数化
df_train['ExerciseAngina'] = df_train['ExerciseAngina'].apply(lambda x: 1 if x == 'Y' else 0)
df_test['ExerciseAngina'] = df_test['ExerciseAngina'].apply(lambda x: 1 if x == 'Y' else 0)

#Oldpeak 0.1以上か否かでダミー変数化
df_train = df_train.drop('Oldpeak',axis=1)
df_test['Oldpeak_dummy'] = df_test['Oldpeak'].apply(lambda x:1 if x >= 0.1 else 0)
df_test = df_test.drop('Oldpeak',axis=1)

#ST_slope ダミー変数化
df_train['ST_Slope_Flat'] = df_train['ST_Slope'].apply(lambda x:1 if x=='Flat' else 0)
df_train['ST_Slope_Down'] = df_train['ST_Slope'].apply(lambda x:1 if x=='Down' else 0)
df_train = df_train.drop('ST_Slope',axis=1)
df_test['ST_Slope_Flat'] = df_test['ST_Slope'].apply(lambda x:1 if x=='Flat' else 0)
df_test['ST_Slope_Down'] = df_test['ST_Slope'].apply(lambda x:1 if x=='Down' else 0)
df_test = df_test.drop('ST_Slope',axis=1)

# MaxHR 欠損か否か 130以下か否かでダミー変数化
df_train['MaxHR_130under'] = df_train['MaxHR'].apply(lambda x:1 if x<= 130 else 0)
df_train = df_train.drop('MaxHR',axis=1)
df_test['MaxHR_nan'] = df_test['MaxHR'].apply(lambda x:1 if np.isnan(x) else 0)
df_test['MaxHR_130under'] = df_test['MaxHR'].apply(lambda x:1 if x<=130 else 0)
df_test = df_test.drop('MaxHR',axis=1)

#データの確認
display(df_train.info())
display(df_test.info())

```

```python
#説明変数と目的変数に分割
X = df_train.drop('HeartDisease',axis=1)
y = df_train['HeartDisease']
```

# 4.予測と評価

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty='l1',solver='liblinear',C=10)
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
y_pred_proba = lr.predict_proba(X_test)[:,1]

#スコア算出
from sklearn.metrics import confusion_matrix, accuracy_score,f1_score,log_loss,roc_auc_score
print('confusion matrix = ¥n',confusion_matrix(y_true=y_test,y_pred=y_pred))
print('accuracy = ',accuracy_score(y_true=y_test,y_pred=y_pred))
print('f1 score = ',f1_score(y_true=y_test,y_pred=y_pred))
print('log_loss = ',log_loss(y_test,y_pred_proba))
print('auc = ',roc_auc_score(y_test,y_pred_proba))
pd.DataFrame(lr.coef_,columns=X.columns).T
```

```python
#決定木分析
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)

from sklearn import tree
clf=tree.DecisionTreeClassifier(max_depth=3) #max_depthを指定
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
y_pred_proba=clf.predict_proba(X_test)[:, 1]

#スコア算出
print('confusion matrix = \n', confusion_matrix(y_true=y_test, y_pred=y_pred))
print('accuracy = ', accuracy_score(y_true=y_test, y_pred=y_pred))
print('f1 score = ', f1_score(y_true=y_test, y_pred=y_pred))
print("log_loss=",log_loss(y_test,y_pred_proba))
print("auc=",roc_auc_score(y_test,y_pred_proba))


#可視化
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(clf, feature_names=X_train.columns, class_names=True, filled=True,fontsize=16);
```
