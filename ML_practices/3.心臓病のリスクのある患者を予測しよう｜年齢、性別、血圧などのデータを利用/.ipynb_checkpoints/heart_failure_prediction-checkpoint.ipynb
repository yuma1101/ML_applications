##機械学習の前処理

```python
#変数の再確認
df_train_master.info()
```

```python
#機械学習用のデータフレームを作成
df_train=df_train_master.copy()
df_submit=df_submit_master.copy()

#ID
df_train=df_train.drop("ID",axis=1)

#Sex 男性を1、女性を0に
df_train["Sex"]=df_train["Sex"].apply(lambda x:1 if x=="M" else 0)
df_submit["Sex"]=df_submit["Sex"].apply(lambda x:1 if x=="M" else 0)

#ChestPainType それぞれのカテゴリーでダミー変数を作る
df_train["ChestPainType_ASY"]=df_train["ChestPainType"].apply(lambda x:1 if x=="ASY" else 0)
df_train["ChestPainType_NAP"]=df_train["ChestPainType"].apply(lambda x:1 if x=="NAP" else 0)
df_train["ChestPainType_ATA"]=df_train["ChestPainType"].apply(lambda x:1 if x=="ATA" else 0)
df_train=df_train.drop("ChestPainType",axis=1)
df_submit["ChestPainType_ASY"]=df_submit["ChestPainType"].apply(lambda x:1 if x=="ASY" else 0)
df_submit["ChestPainType_NAP"]=df_submit["ChestPainType"].apply(lambda x:1 if x=="NAP" else 0)
df_submit["ChestPainType_ATA"]=df_submit["ChestPainType"].apply(lambda x:1 if x=="ATA" else 0)
df_submit=df_submit.drop("ChestPainType",axis=1)

#RestingBP nullのデータを削除
df_train=df_train.dropna(subset=["RestingBP"])

#Cholesterol trainデータの平均値で補完(df_submitもtrainデータの平均値で補完)
df_train["Cholesterol"]=df_train["Cholesterol"].fillna(np.mean(df_train["Cholesterol"]))
df_submit["Cholesterol"]=df_submit["Cholesterol"].fillna(np.mean(df_train["Cholesterol"]))

#RestingECG ダミー変数化
df_train["RestingECG_ST"]=df_train["RestingECG"].apply(lambda x:1 if x=="ST" else 0)
df_train=df_train.drop("RestingECG",axis=1)
df_submit["RestingECG_ST"]=df_submit["RestingECG"].apply(lambda x:1 if x=="ST" else 0)
df_submit=df_submit.drop("RestingECG",axis=1)

#ExerciseAngina ダミー変数化
df_train["ExerciseAngina"]=df_train["ExerciseAngina"].apply(lambda x:1 if x=="Y" else 0)
df_submit["ExerciseAngina"]=df_submit["ExerciseAngina"].apply(lambda x:1 if x=="Y" else 0)

#Oldpeak 0.1以上か否かでダミー変数化(df_trainはEDA時に作成済)
df_train=df_train.drop("Oldpeak",axis=1)
df_submit["Oldpeak_dummy"]=df_submit["Oldpeak"].apply(lambda x:1 if x>=0.1 else 0)
df_submit=df_submit.drop("Oldpeak",axis=1)

#ST_slope ダミー変数化
df_train["ST_Slope_Flat"]=df_train["ST_Slope"].apply(lambda x:1 if x=="Flat" else 0)
df_train["ST_Slope_Down"]=df_train["ST_Slope"].apply(lambda x:1 if x=="Down" else 0)
df_train=df_train.drop("ST_Slope",axis=1)
df_submit["ST_Slope_Flat"]=df_submit["ST_Slope"].apply(lambda x:1 if x=="Flat" else 0)
df_submit["ST_Slope_Down"]=df_submit["ST_Slope"].apply(lambda x:1 if x=="Down" else 0)
df_submit=df_submit.drop("ST_Slope",axis=1)

#MaxHR 欠損か否か、130以下か否かでダミー変数化
df_train["MaxHR_130under"]=df_train["MaxHR"].apply(lambda x:1 if x<=130 else 0)
df_train=df_train.drop("MaxHR",axis=1)
df_submit["MaxHR_nan"]=df_submit["MaxHR"].apply(lambda x:1 if np.isnan(x) else 0)
df_submit["MaxHR_130under"]=df_submit["MaxHR"].apply(lambda x:1 if x<=130 else 0)
df_submit=df_submit.drop("MaxHR",axis=1)

#データの確認
display(df_train.info())
display(df_submit.info())
```

```python
#説明変数と目的変数に分割
X=df_train.drop("HeartDisease",axis=1)
y=df_train["HeartDisease"]
```

##予測と評価

```python
#ロジスティック回帰(sklearn)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty="l1",solver="liblinear",C=10) #C値で正則化の強さを調整(低いほど強い)
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)
y_pred_proba=lr.predict_proba(X_test)[:, 1]

#スコア算出
from sklearn.metrics import confusion_matrix, accuracy_score,f1_score,log_loss,roc_auc_score
print('confusion matrix = \n', confusion_matrix(y_true=y_test, y_pred=y_pred))
print('accuracy = ', accuracy_score(y_true=y_test, y_pred=y_pred))
print('f1 score = ', f1_score(y_true=y_test, y_pred=y_pred))
print("log_loss=",log_loss(y_test,y_pred_proba))
print("auc=",roc_auc_score(y_test,y_pred_proba))
pd.DataFrame(lr.coef_,columns=X.columns).T
```

```python
#標準化偏回帰係数を確認

#標準化
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_std=ss.fit_transform(X)
X_std=pd.DataFrame(X_std,columns=X.columns)

#ロジスティック回帰(sklearn)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std,y,test_size=0.2,random_state=1)

from sklearn.linear_model import LogisticRegression
lr2 = LogisticRegression(penalty="l1",solver="liblinear",C=10) #C値で正則化の強さを調整(低いほど強い)
lr2.fit(X_train, y_train)
y_pred=lr2.predict(X_test)
y_pred_proba=lr2.predict_proba(X_test)[:, 1]

#標準化偏回帰係数を算出
pd.DataFrame(lr2.coef_,columns=X_std.columns).T

#MaxHR_nan、ST_slope_Flat、ST_slope_Downの順に心臓病に影響している
```

```python
#決定木分析
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)

from sklearn import tree
clf=tree.DecisionTreeClassifier(max_depth=3) #max_depthを指定
clf=clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
y_pred_proba=clf.predict_proba(X_test)[:, 1]

#スコア算出
print('confusion matrix = \n', confusion_matrix(y_true=y_test, y_pred=y_pred))
print('accuracy = ', accuracy_score(y_true=y_test, y_pred=y_pred))
print('f1 score = ', f1_score(y_true=y_test, y_pred=y_pred))
print("log_loss=",log_loss(y_test,y_pred_proba))
print("auc=",roc_auc_score(y_test,y_pred_proba))


#可視化
from sklearn.tree import plot_tree
plt.figure(figsize=(20,20))
plot_tree(clf, feature_names=X_train.columns, class_names=True, filled=True,fontsize=16);
```
