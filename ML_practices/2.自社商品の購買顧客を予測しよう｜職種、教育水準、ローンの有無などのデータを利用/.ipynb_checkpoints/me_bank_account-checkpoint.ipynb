# 2.6balance:年間残高

```python
fig = plt.subplots(figsize=(20,10))
sns.histplot(train[train['y'] == 'no']['balance'],kde=False,label='no',color="blue")
sns.histplot(train[train['y'] == 'yes']['balance'],kde=False,label='yes',color="yellow")
plt.legend()
```

```python
#外れ値の対処法として対数変換がある
#絶対値の対数をとってから元の符号を付加する(対象が整数かつ負の数を含む連続変数の場合のみ使用可能な方法)
fig = plt.subplots(figsize=(20,10))
sns.histplot(train[train['y'] == 'no']['balance'].map(lambda x: np.sign(x) * np.log(np.abs(x))),label="no")
sns.histplot(train[train['y'] == 'yes']['balance'].map(lambda x: np.sign(x) * np.log(np.abs(x))),label="yes")

# 絶対値の対数を取ってから元の符号を付加する(対象が整数かつ負の数を含む連続変数の場合のみ使用可能な方法)
# fig= plt.subplots(figsize=(20,10))
# sns.distplot(train[train['y']=='no']['balance'].map(lambda x: np.sign(x) * np.log(np.abs(x))), label="no")
# sns.distplot(train[train['y']=='yes']['balance'].map(lambda x: np.sign(x) * np.log(np.abs(x))), label="yes")
# plt.legend();
#残高が多いほどyesになる傾向
```

## 3.1関数定義

```python
def submit(df,pred):
    df['y'] = pred
    df = pd.concat([df['ID'],df['y']], axis=1)
    df = df.set_index('ID')
    return df 
```

## 3.2 前処理

### 3.2.1 前処理

```python
# year:年を追加
df.loc[0:27728,'year'] = int(2008)
df.loc[27729:41590, 'year'] = int(2009)
df.loc[42591:45210, 'year'] = int(2010)

# allデータからtrainとtestに分割
train = df[df['y'].notnull()]
test = df[df['y'].isnull()].drop('y',axis=1)
```
