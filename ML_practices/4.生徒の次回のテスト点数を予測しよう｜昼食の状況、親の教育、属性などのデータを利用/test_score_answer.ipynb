```python
#reading score
fig = plt.subplots(figsize=(10,10))
sns.regplot(data=df_train_master, x="reading score", y="math score" , line_kws={"color": "black"})

#相関係数の算出＋検定の実施
from scipy import stats
stats.pearsonr(df_train_master["reading score"], df_train_master["math score"])
#reading scoreとmath scoreには強い正の相関関係を認める
```

```python
#writing score(欠損あり)
fig = plt.subplots(figsize=(10,10))
sns.regplot(data = df_train_master, x="writing score", y="math score" , line_kws={"color": "black"})

#相関係数の算出＋検定の実施
from scipy import stats
stats.pearsonr(df_train_master["writing score"].fillna(np.mean(df_train_master["writing score"])), df_train_master["math score"])
#writing scoreとmath scoreには強い正の相関関係を認める
```

```python
#reading scoreとwriting scoreの関係性
fig = plt.subplots(figsize=(10,10))
sns.regplot(data=df_train_master, x="writing score", y="reading score", line_kws={"color": "black"})

#相関係数の算出＋検定の実施
from scipy import stats
stats.pearsonr(df_train_master["writing score"].fillna(np.mean(df_train_master["writing score"])), df_train_master["reading score"])
#reading scoreとwriting scoreは非常に強い相関関係を持つ
#多重共線性も考慮して、reading scoreのみ、またはreading scoreとwriting scoreの平均値のみを使うのが良さそう
```

```python
#reading scoreとwriting scoreの平均値とmath scoreの相関関係を検証
x = df_train_master.copy()
x["avg score"] = (df_train_master["reading score"] + df_train_master["writing score"]) / 2
fig = plt.subplots(figsize=(10,10))
sns.regplot(data=x, x="avg score", y="math score", line_kws={"color": "black"})

#相関係数の算出＋検定の実施
from scipy import stats
stats.pearsonr(np.where(x["avg score"] == x["avg score"], x["avg score"], x["reading score"]), x["math score"])
#reading score単体よりも平均を使った方が相関係数が少しだけ高いため、今回はwriting scoreとreading scoreの平均値をscoreとして使用する
```

```python
#機械学習用のデータフレームを作成
df_train = df_train_master.copy()
df_submit = df_submit_master.copy()

#IDは削除
df_train = df_train.drop("ID",axis=1)

#gender 男性を1、女性を0に
df_train["gender"] = df_train["gender"].apply(lambda x: 1 if x == "male" else 0)
df_submit["gender"] = df_submit["gender"].apply(lambda x: 1 if x == "male" else 0)

#race/ethnicity それぞれのカテゴリーでダミー化
df_train["race/ethnicity_A"] = df_train["race/ethnicity"].apply(lambda x: 1 if x == "group A" else 0)
df_train["race/ethnicity_B"] = df_train["race/ethnicity"].apply(lambda x: 1 if x == "group B" else 0)
df_train["race/ethnicity_C"] = df_train["race/ethnicity"].apply(lambda x: 1 if x == "group C" else 0)
df_train["race/ethnicity_D"] = df_train["race/ethnicity"].apply(lambda x: 1 if x == "group D" else 0)
df_train = df_train.drop("race/ethnicity",axis=1)
df_submit["race/ethnicity_A"] = df_submit["race/ethnicity"].apply(lambda x: 1 if x == "group A" else 0)
df_submit["race/ethnicity_B"] = df_submit["race/ethnicity"].apply(lambda x: 1 if x == "group B" else 0)
df_submit["race/ethnicity_C"] = df_submit["race/ethnicity"].apply(lambda x: 1 if x == "group C" else 0)
df_submit["race/ethnicity_D"] = df_submit["race/ethnicity"].apply(lambda x: 1 if x == "group D" else 0)
df_submit = df_submit.drop("race/ethnicity",axis=1)

#parental level of education　それぞれのカテゴリーでダミー化
df_train["parental level of education_some_high"] = df_train["parental level of education"].apply(lambda x: 1 if x == "some high school" else 0)
df_train["parental level of education_associate"] = df_train["parental level of education"].apply(lambda x: 1 if x == "associate's degree" else 0)
df_train["parental level of education_some_college"] = df_train["parental level of education"].apply(lambda x: 1 if x == "some college" else 0)
df_train["parental level of education_master"] = df_train["parental level of education"].apply(lambda x: 1 if x == "master's degree" else 0)
df_train["parental level of education_bachelor"] = df_train["parental level of education"].apply(lambda x: 1 if x == "bachelor's degree" else 0)
df_train = df_train.drop("parental level of education",axis=1)
df_submit["parental level of education_some_high"] = df_submit["parental level of education"].apply(lambda x: 1 if x == "some high school" else 0)
df_submit["parental level of education_associate"] = df_submit["parental level of education"].apply(lambda x: 1 if x == "associate's degree" else 0)
df_submit["parental level of education_some_college"] = df_submit["parental level of education"].apply(lambda x: 1 if x == "some college" else 0)
df_submit["parental level of education_master"] = df_submit["parental level of education"].apply(lambda x: 1 if x == "master's degree" else 0)
df_submit["parental level of education_bachelor"] = df_submit["parental level of education"].apply(lambda x: 1 if x == "bachelor's degree" else 0)
df_submit = df_submit.drop("parental level of education",axis=1)

#lunch standardを1に
df_train["lunch"] = df_train["lunch"].apply(lambda x: 1 if x == "standard" else 0)
df_submit["lunch"] = df_submit["lunch"].apply(lambda x: 1 if x == "standard" else 0)

#test preparation course completedを1に
df_train["test preparation course"] = df_train["test preparation course"].apply(lambda x: 1 if x == "completed" else 0)
df_submit["test preparation course"] = df_submit["test preparation course"].apply(lambda x: 1 if x == "completed" else 0)

#reading score 0のデータを削除
df_train = df_train.query('`reading score` != 0')

#writing score　欠損値にreading scoreを代入し、平均を計算してavg scoreを作成(reading scoreとwriting scoreは削除)
df_train["writing score"] = np.where(df_train["writing score"] == df_train["writing score"], df_train["writing score"], df_train["reading score"])
df_train["avg score"] = (df_train["reading score"] + df_train["writing score"]) / 2
df_train = df_train.drop(["writing score", "reading score"], axis=1)
df_submit["writing score"] = np.where(df_submit["writing score"] == df_submit["writing score"], df_submit["writing score"], df_submit["reading score"])
df_submit["avg score"] = (df_submit["reading score"] + df_submit["writing score"]) / 2
df_submit = df_submit.drop(["writing score", "reading score"], axis=1)

#データの確認
display(df_train.info())
display(df_train.describe())
display(df_submit.info())
display(df_submit.describe())
```

```python
#説明変数と目的変数に分割
X = df_train.drop("math score", axis=1)
y = df_train["math score"]
```

##予測と評価

```python
#重回帰(sklearn)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

from sklearn.linear_model import LinearRegression
lr = LinearRegression() 
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

#スコア算出
from sklearn.metrics import r2_score,mean_squared_error
print("r2=", r2_score(y_true=y_test, y_pred=y_pred))
print("RMSE=", np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_pred)))
pd.DataFrame(lr.coef_, index=X.columns)
```

```python
#標準化偏回帰係数を確認

#標準化
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_std = ss.fit_transform(X)
X_std = pd.DataFrame(X_std,columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2, random_state=2)
lr2 = LinearRegression() 
lr2.fit(X_train, y_train)
y_pred = lr2.predict(X_test)

#標準化偏回帰係数を算出
pd.DataFrame(lr2.coef_, index=X_std.columns)
#avg score,genderがscoreの予測に寄与
```

```python
#決定木分析
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=3) #max_depthを指定
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)

#スコア算出
from sklearn.metrics import r2_score,mean_squared_error
print("r2=", r2_score(y_true=y_test, y_pred=y_pred))
print("RMSE=", np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_pred)))
pd.DataFrame(lr.coef_, index=X.columns)

#可視化
from sklearn.tree import plot_tree
plt.figure(figsize=(30,20))
plot_tree(clf, feature_names=X_train.columns, class_names=True, filled=True,fontsize=16);
```

```python
#ベースラインモデル math scoreの平均値を代入して予測
y_pred = [np.mean(y_train)] * 180
print("RMSE=", np.sqrt(mean_squared_error(y_true=y_test, y_pred=y_pred)))
#重回帰モデルのスコアがベースラインモデルを上回っていることを確認
```
