# 3.機械学習の前処理

```python
# 機械学習用のデータフレームを作成
df_train = train.copy()
df_test = test.copy()

# indexは削除
df_train = df_train.drop("index",axis=1)

# gender 欠損の削除 + Maleを1、Femaleを0にダミー変数化
df_train = df_train.query('gender  == gender')
df_train["gender"] = df_train["gender"].apply(lambda x : 1 if x == "Male" else 0)
df_test["gender"] = df_test["gender"].apply(lambda x: 1 if x == "Male" else 0)

# age 10歳区切りで順序尺度に変換
df_train["age"] = df_train["age"] // 10
df_test["age"] = df_test["age"] // 10

# hypertensiion 処理なし
# heart_desease 処理なし

# smoking_history 欠損を1つのカテゴリーとして扱い、one-hot encodingでそれぞれのカテゴリーをダミー変数(baseはnever)
df_train["smoking_history_NULL"] = df_train["smoking_history"].apply(lambda x :1 if x != x else 0)
df_train["smoking_history_ever"] = df_train["smoking_history"].apply(lambda x :1 if x == "ever" else 0)
df_train["smoking_history_current"] = df_train["smoking_history"].apply(lambda x :1 if x == "current" else 0)
df_train["smoking_hitory_not_current"] = df_train["smoking_history"].apply(lambda x: 1 if x == "not current" else 0)
df_train["smoking_history_former"] = df_train["smoking_history"].apply(lambda x : 1 if x == "former" else 0)
df_train = df_train.drop("smoking_history" ,axis=1)
df_test["smoking_history_NULL"] = df_test["smoking_history"].apply(lambda x :1 if x != x else 0)
df_test["smoking_history_ever"] = df_test["smoking_history"].apply(lambda x :1 if x == "ever" else 0)
df_test["smoking_history_current"] = df_test["smoking_history"].apply(lambda x :1 if x == "current" else 0)
df_test["smoking_hitory_not_current"] = df_test["smoking_history"].apply(lambda x: 1 if x == "not current" else 0)
df_test["smoking_history_former"] = df_test["smoking_history"].apply(lambda x : 1 if x == "former" else 0)
df_test = df_test.drop("smoking_history" ,axis=1)

# bmi 0は中央値で補完
df_train["bmi"] = df_train["bmi"].fillna(df_train["bmi"].median())
df_test["bmi"] = df_test["bmi"].fillna(df_train["bmi"].median())

#HbA1c_level 「5.5以下か否か」、「7.0以上か否か」のダミー変数化
df_train["HbA1c_5.5under"] = df_train["HbA1c_level"].apply(lambda x : 1 if x <= 5.5 else 0)
df_train["HbA1c_7.0over"] = df_train["HbA1c_level"].apply(lambda x : 1 if x >= 7.0 else 0)
df_train = df_train.drop("HbA1c_level",axis=1)
df_test["HbA1c_5.5under"] = df_test["HbA1c_level"].apply(lambda x : 1 if x <= 5.5 else 0)
df_test["HbA1c_7.0over"] = df_test["HbA1c_level"].apply(lambda x : 1 if x >= 7.0 else 0)
df_test = df_test.drop("HbA1c_level",axis=1)

# blood_glcose_level 「126より小さいか否か」「200より小大きいか否か」のダミー変数化
df_train["glucose_126under"] = df_train["blood_glucose_level"].apply(lambda x : 1 if x < 126 else 0)
df_train["glucose_200over"]  = df_train["blood_glucose_level"].apply(lambda x : 1 if x > 200 else 0)
df_train = df_train.drop("blood_glucose_level",axis=1)
df_test["glucose_126under"] = df_test["blood_glucose_level"].apply(lambda x : 1 if x < 126 else 0)
df_test["glucose_200over"]  = df_test["blood_glucose_level"].apply(lambda x : 1 if x > 200 else 0)
df_test = df_test.drop("blood_glucose_level",axis=1)

# データの確認
display(df_train.info())
display(df_train.describe())

display(df_test.info())
display(df_test.describe())

```

```python
# heatmapを作成する(文字列が含まれているためエラーが出る)
# fig = plt.subplots(figsize=(20,10))
# sns.heatmap(df_train.corr(),cmap="bwr",annot=True,vmax=1,vmin=-1,center=0)

# ✅ 方法①：数値列だけで相関を取る（最も一般的）
df_corr = df_train.select_dtypes(include="number").corr()
fig = plt.subplots(figsize=(20,10))
sns.heatmap(df_corr, cmap="bwr", annot=True, vmax=1, vmin=-1, center=0)
```

# 4.予測と評価

```python
# 説明変数と目的変数に分割
X = df_train.drop("diabetes",axis=1)
y = df_train["diabetes"]
```

```python
# ロジスティック回帰(sklearn)
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty="l1",solver="liblinear",C=1)# C値で正規化の強さを調整(低いほど強い)  
lr.fit(X_train,y_train)
y_pred_proba = lr.predict_proba(X_test)[:, 1]
y_pred = np.where(y_pred_proba > 0.5, 1, 0)

# スコア算出
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score,log_loss
print('confusion matrix = \n', confusion_matrix(y_true=y_test, y_pred=y_pred))
print('accuracy = ', accuracy_score(y_true=y_test, y_pred=y_pred))
print('recall = ', recall_score(y_true=y_test, y_pred=y_pred))
print('precision = ', precision_score(y_true=y_test, y_pred=y_pred))
print('log_loss = ', log_loss(y_true=y_test, y_pred=y_pred_proba))
pd.DataFrame(lr.coef_, columns=X.columns)
```

```python
# 標準化回帰係数を確認
# 標準化
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_std = ss.fit_transform(X)
X_std = pd.DataFrame(X_std, columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2, random_state=1)
lr2 = LogisticRegression(penalty="l1", solver="liblinear",C=1)
lr2.fit(X_train,y_train)
y_pred = lr2.predict(X_test)
pd.DataFrame(lr2.coef_, columns=X_std.columns).T
```

```python
# 決定木分析
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)

from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=8, random_state=1)# max_depthを指定
clf = clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)  
y_pred_proba = clf.predict_proba(X_test)[:, 1]

# スコア算出
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, log_loss
print('confusion matrix = \n', confusion_matrix(y_true=y_test, y_pred=y_pred))
print('accuracy = ', accuracy_score(y_true=y_test, y_pred=y_pred))
print('recall = ', recall_score(y_true=y_test, y_pred=y_pred))
print('precision = ', precision_score(y_true=y_test, y_pred=y_pred))
print('log_loss = ', log_loss(y_test, y_pred_proba))

# 可視化
from sklearn.tree import plot_tree
plt.figure(figsize=(60,20))
plot_tree(clf, feature_names=X_train.columns, class_names=True, filled=True, fontsize=16)

# 分類にメインで使用している特徴量はロジスティックとほぼ一緒(本当はHbA1cやglucose、ageは変換前の連続ち値を使用した方が酔いが、ややこしいのでここでは回帰分析の特徴量をそのまま使用)
# ロジスティック回帰の方がスコアが良いのでそちらを採用
```

```python
# ベースラインモデル 全て0で予測(ベースラインモデルとは「何もしなくてもこの程度は当たる」基準モデル)
y_pred = [0] * len(y_test)
print('accuracy = ', accuracy_score(y_true=y_test, y_pred=y_pred))

# accuracyがベースラインモデルを上回っていることを確認
```
