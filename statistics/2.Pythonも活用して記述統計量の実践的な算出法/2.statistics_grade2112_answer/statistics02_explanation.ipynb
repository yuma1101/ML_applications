# 問2 Pythonを使って記述統計量の算出

```python
import numpy as np
import pandas as pd
```

```python
data = [45, 40, 55, 60, 65, 35]
df = pd.DataFrame(data)
```

## 1.平均値

```python
np.mean(data)
```

```python
df.mean()
```

## 2.偏差の和

```python
(45 - 50) + (40 - 50) + (55 - 50) + (60 - 50) + (65 - 50) + (35 - 50) 
```

## 3.平方和

```python
l = []
for x in data:
    l.append((x - np.mean(data))**2)
sum(l)
```

```python
l = [(x - df.mean())**2 for x in data]
np.sum(l)
```

```python
sum(np.square(data)) - (sum(data))**2 / len(data)
```

## 4.分散

```python
np.var(data)
```

```python
df.var(ddof=0)
```

# 問5 尖度と歪度

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes, load_linnerud, fetch_california_housing
```

```python
data1 = fetch_california_housing()
df1 = pd.DataFrame(data1.data, columns=data1.feature_names)
```

```python
plt.figure(figsize=(10, 5))
for i, col in enumerate(df1.columns):
    plt.subplot(3, 3, i+1)
    sns.kdeplot(df1[col])
plt.tight_layout()
plt.show()
```

```python
data2 = load_diabetes()
df2 = pd.DataFrame(data2.data, columns=data2.feature_names)
```

```python
plt.figure(figsize=(10, 5))
for i, col in enumerate(df2.columns):
    plt.subplot(3, 4, i+1)
    sns.kdeplot(df2[col])
plt.tight_layout()
plt.show()
```
