```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

### 二項分布

```python
n = 10
p = 0.2
```

```python
x = np.arange(10)
```

```python
y = stats.binom.pmf(x, n, p)
```

```python
plt.bar(x, y)
plt.show()
```

### ポアソン分布

```python
lambda_ = 2
```

```python
x = np.arange(10)
```

```python
y = stats.poisson.pmf(x, lambda_)
```

```python
plt.bar(x, y)
plt.show()
```

### 幾何分布

```python
p = 0.2
```

```python
x = np.arange(10)
```

```python
y = stats.geom.pmf(x, p)
```

```python
plt.bar(x, y)
plt.show()
```

### 指数分布

```python
lambda_ = 0.2
```

```python
x = np.arange(10)
```

```python
y = stats.expon.pdf(x, scale=1/lambda_)
```

```python
plt.plot(x, y)
plt.show()
```

### 超幾何分布

```python
N = 10
M = 6
n = 3
```

```python
x = np.arange(10)
```

```python
y = stats.hypergeom.pmf(x, N, M, n)
```

```python
plt.bar(x, y)
plt.show()
```

### 一様分布

```python
x = np.linspace(-2, 4, 1001)
```

```python
y = stats.uniform.pdf(x, loc=0, scale=1)
```

```python
plt.plot(x, y)
plt.show()
```
