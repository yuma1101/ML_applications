# 問2 Pythonを使って母平均の検定（母分散未知）

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data = [10.3, 10.5, 9.8, 10.2, 10.4, 10.8, 10.7, 10.7, 9.7, 9.9]
```

```python
mu0 = 10.0
```

```python
t0, p = stats.ttest_1samp(data, mu0, alternative="two-sided")
t0, p
```

p値 < 有意水準0.05　なので帰無仮説H0を棄却する

グラフに描画

```python
x = np.linspace(-5, 5, 101)
y = stats.t.pdf(x, len(data)-1)
```

```python
alpha = 0.05
```

```python
lower_cv = stats.t.ppf(alpha/2, len(data)-1)
upper_cv = stats.t.ppf(1-alpha/2, len(data)-1)
(lower_cv, upper_cv)
```

```python
plt.plot(x, y)
plt.vlines(x=t0, ymin=0, ymax=0.2, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(-5, lower_cv, 100)
plt.fill_between(lower_x, stats.t.pdf(lower_x, len(data)-1), facecolor='orange')

upper_x = np.linspace(upper_cv, 5, 100)
plt.fill_between(upper_x, stats.t.pdf(upper_x, len(data)-1), facecolor='orange')
plt.grid()
plt.show()
```

# 問4 Pythonを使って母分散の検定

```python
data = [10, 9, 9, 9, 10, 10, 11, 9, 9, 9]
```

```python
alpha = 0.05
```

```python
sigma0_2 = 1.5 #母分散
s = np.var(data, ddof=1) #不偏分散
dof = len(data) - 1 #自由度
S = s * dof  #平方和
```

```python
chi0_2 = S / sigma0_2
chi0_2
```

```python
lower_cv = stats.chi2.ppf(alpha, dof)
lower_cv
```

chi0_2 < lower_cv　なので帰無仮説H0を棄却する

グラフに描画

```python
x = np.linspace(0, 30, 101)
y = stats.chi2.pdf(x, dof)
```

```python
plt.plot(x, y)
plt.vlines(x=chi0_2, ymin=0, ymax=0.1, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(0, lower_cv, 100)
plt.fill_between(lower_x, stats.chi2.pdf(lower_x, dof), facecolor='orange')
plt.grid()
plt.show()
```

# 問6 Pythonを使って母比率の検定

```python
p0 = 0.11
```

```python
n = 200 #サンプル数
p_hat = 20 / n
```

```python
alpha = 0.05
```

```python
z0 = (p_hat - p0) / np.sqrt(p0*(1 - p0) / n)
z0
```

```python
lower_cv = stats.norm.ppf(alpha/2)
upper_cv = stats.norm.ppf(1-alpha/2)
(lower_cv, upper_cv)
```

lower_cv < z0 < upper_cv　なので帰無仮説H0を棄却できない

グラフに描画

```python
x = np.linspace(-5, 5, 101)
y = stats.norm.pdf(x)
```

```python
plt.plot(x, y)
plt.vlines(x=z0, ymin=0, ymax=0.4, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(-5, lower_cv, 100)
plt.fill_between(lower_x, stats.norm.pdf(lower_x), facecolor='orange')

upper_x = np.linspace(upper_cv, 5, 100)
plt.fill_between(upper_x, stats.norm.pdf(upper_x), facecolor='orange')
plt.grid()
plt.show()
```

# 問8 Pythonを使って母平均の差の検定（対応のあるデータ）

```python
data1 = [8.0, 5.4, 4.8, 5.8, 6.4, 5.8]
data2 = [7.0, 5.0, 3.6, 7.0, 7.0, 5.4]
```

```python
t0, p = stats.ttest_rel(data1, data2, alternative="two-sided")
t0, p
```

p値 > 有意水準0.05　なので帰無仮説H0は棄却できない

```python
data = [round(x1 - x2, 1) for x1, x2 in zip(data1, data2)]
data
```

```python
alpha = 0.05
```

```python
xd_bar = np.mean(data)
sd_2 = np.var(data, ddof=1)
dof = len(data) - 1
```

```python
t0 = xd_bar / np.sqrt(sd_2/len(data))
t0
```

```python
lower_cv = stats.t.ppf(alpha/2, df=dof)
upper_cv = stats.t.ppf(1-alpha/2, df=dof)
(lower_cv, upper_cv)
```

lower_cv < t0 < upper_cv　なので帰無仮説H0を棄却できない

グラフに描画

```python
x = np.linspace(-5, 5, 101)
y = stats.t.pdf(x, df=dof)
```

```python
plt.plot(x, y)
plt.vlines(x=t0, ymin=0, ymax=0.4, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(-5, lower_cv, 100)
plt.fill_between(lower_x, stats.t.pdf(lower_x, df=dof), facecolor='orange')

upper_x = np.linspace(upper_cv, 5, 100)
plt.fill_between(upper_x, stats.t.pdf(upper_x, df=dof), facecolor='orange')
plt.grid()
plt.show()
```

# 問12 Pythonを使って母平均の差の検定（対応のないデータ、分散未知だが等しい場合）

```python
data1 = [8.0, 5.4, 4.8, 5.8, 6.4, 5.6]
data2 = [9.0, 7.0, 3.6, 6.4, 6.0]
```

```python
t0, p = stats.ttest_ind(data1, data2, equal_var=True, alternative="two-sided")
t0, p
```

p値 > 有意水準0.05　なので帰無仮説H0は棄却できない

```python
alpha = 0.05
```

```python
x_bar1 = np.mean(data1)
x_bar2 = np.mean(data2)
```

```python
s1_2 = np.var(data1, ddof=1)
s2_2 = np.var(data2, ddof=1)
s1_2, s2_2
```

```python
dof1 = len(data1) - 1
dof2 = len(data2) - 1
```

```python
s_2 = (dof1 * s1_2 + dof2 * s2_2) / (dof1 + dof2) #自由度で重み付けした不偏分散の平均
s_2
```

```python
t0 = (x_bar1 - x_bar2) / np.sqrt(s_2/len(data1) + s_2/len(data2))
t0
```

```python
lower_cv = stats.t.ppf(alpha/2, df=dof1+dof2)
upper_cv = stats.t.ppf(1-alpha/2, df=dof1+dof2)
(lower_cv, upper_cv)
```

lower_cv < t0 < upper_cv　なので帰無仮説H0を棄却できない

グラフに描画

```python
x = np.linspace(-5, 5, 101)
y = stats.t.pdf(x, df=dof1+dof2)
```

```python
plt.plot(x, y)
plt.vlines(x=t0, ymin=0, ymax=0.4, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(-5, lower_cv, 100)
plt.fill_between(lower_x, stats.t.pdf(lower_x, df=dof1+dof2), facecolor='orange')

upper_x = np.linspace(upper_cv, 5, 100)
plt.fill_between(upper_x, stats.t.pdf(upper_x, df=dof1+dof2), facecolor='orange')
plt.grid()
plt.show()
```

# 問16 Pythonを使って母分散の比の検定

```python
dataA = [10, 9, 8, 9, 10]
dataB = [10, 10, 9, 9, 9, 10]
```

```python
alpha = 0.05
```

```python
sA_2 = np.var(dataA, ddof=1)
sB_2 = np.var(dataB, ddof=1)
sA_2, sB_2
```

```python
dofA = len(dataA) - 1
dofB = len(dataB) - 1
```

```python
lower_cv = stats.f.ppf(alpha/2, dofA, dofB)
upper_cv = stats.f.ppf(1-alpha/2, dofA, dofB)
(lower_cv, upper_cv)
```

```python
F0 = sA_2 / sB_2
F0
```

lower_cv < F0 < upper_cv　なので帰無仮説H0を棄却できない

グラフに描画

```python
x = np.linspace(0, 10, 101)
y = stats.f.pdf(x, dofA, dofB)
```

```python
plt.plot(x, y)
plt.vlines(x=F0, ymin=0, ymax=0.4, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(0, lower_cv, 100)
plt.fill_between(lower_x, stats.f.pdf(lower_x, dofA, dofB), facecolor='orange')

upper_x = np.linspace(upper_cv, 10, 100)
plt.fill_between(upper_x, stats.f.pdf(upper_x, dofA, dofB), facecolor='orange')
plt.grid()
plt.show()
```

```python
stats.f.cdf(F0, dofA, dofB)
```

# 問18 Pythonを使って母比率の差の検定

```python
pA_hat = 20 / 200
pB_hat = 11 / 100
```

```python
nA = 200 
nB = 100
```

```python
alpha = 0.05
```

```python
p = (nA*pA_hat + nB*pB_hat) / (nA + nB) #加重平均
```

```python
z0 = (pA_hat - pB_hat) / np.sqrt(p*(1 - p)*(1/nA + 1/nB))
z0
```

```python
lower_cv = stats.norm.ppf(alpha/2)
upper_cv = stats.norm.ppf(1-alpha/2)
(lower_cv, upper_cv)
```

lower_cv < z0 < upper_cv　なので帰無仮説H0を棄却できない

グラフに描画

```python
x = np.linspace(-5, 5, 101)
y = stats.norm.pdf(x)
```

```python
plt.plot(x, y)
plt.vlines(x=z0, ymin=0, ymax=0.4, color='gray', ls='--', linewidth=2)

lower_x = np.linspace(-5, lower_cv, 100)
plt.fill_between(lower_x, stats.norm.pdf(lower_x), facecolor='orange')

upper_x = np.linspace(upper_cv, 5, 100)
plt.fill_between(upper_x, stats.norm.pdf(upper_x), facecolor='orange')
plt.grid()
plt.show()
```
