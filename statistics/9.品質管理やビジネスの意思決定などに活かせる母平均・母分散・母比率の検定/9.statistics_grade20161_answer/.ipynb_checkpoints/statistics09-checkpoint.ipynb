### ■ 問4　母分散の検定

```python
import numpy as np
from scipy.stats import chi2

# データ
data = np.array([10, 9, 9, 9, 10, 10, 11, 9, 9, 9])
n = len(data)

# 母分散の帰無仮説
sigma0_sq = 1.5

# 標本分散（不偏分散）
s2 = np.var(data, ddof=1)

# χ² 検定統計量
chi2_stat = (n - 1) * s2 / sigma0_sq

# 左片側検定 → χ²分布の左側5%点
alpha = 0.05
critical_value = chi2.ppf(alpha, df=n-1)

# P値
p_value = chi2.cdf(chi2_stat, df=n-1)

print("標本分散 s^2 =", s2)
print("検定統計量 χ² =", chi2_stat)
print("臨界値（5%点） =", critical_value)
print("p値 =", p_value)

if chi2_stat < critical_value:
	print("→ 帰無仮説を棄却：バラつきは小さくなった（従来より改善）")
else:
	print("→ 帰無仮説を棄却できない：バラつきが小さくなったとは言えない")
```

### ■ 問6　母比率の検定

```python
import numpy as np
from math import sqrt
from scipy.stats import norm

# ---- データ ----
n = 200        # 標本数
x = 20         # 不良品数
p0 = 0.11      # 帰無仮説の不良率

# 標本比率
phat = x / n

# ---- Z検定 ----
# 標準誤差
se = sqrt(p0 * (1 - p0) / n)

# Z統計量
z = (phat - p0) / se

# 両側検定の p値
p_value = 2 * (1 - norm.cdf(abs(z)))

# 臨界値（±1.96 が 5%両側の値）
z_critical = norm.ppf(1 - 0.05/2)

print("標本比率 p̂ =", phat)
print("Z 統計量 =", z)
print("p値 =", p_value)
print("臨界値（±） =", z_critical)

# 判定
if abs(z) > z_critical:
    print("⇒ 帰無仮説を棄却：不良率は変わった（統計的に有意）")
else:
    print("⇒ 帰無仮説を棄却できない：不良率が変わったとは言えない")

```

### ■ 問8　母平均の差の検定(対応のあるデータ)

```python
import numpy as np
from scipy.stats import ttest_rel

# データ（前期・後期）
before = np.array([8.0, 5.4, 4.8, 5.8, 6.4, 5.8])
after  = np.array([7.0, 5.0, 3.6, 7.0, 7.0, 5.4])

# 対応のある t 検定
t_stat, p_value = ttest_rel(before, after)

print("t 統計量 =", t_stat)
print("p 値 =", p_value)

# 判定（両側検定）
alpha = 0.05
if p_value < alpha:
    print("⇒ 帰無仮説を棄却：前期と後期の平均は有意に異なる")
else:
    print("⇒ 帰無仮説を棄却できない：前期と後期の平均に有意な差はない")
```

### ■ 問10　母平均の差の検定(対応のないデータ、分散既知)


```python
import numpy as np
from scipy.stats import norm

# -- データ ---
group1 = np.array([8.0, 5.4, 4.8, 5.8, 6.4, 5.6])
group2 = np.array([9.0, 7.0, 3.6, 6.4, 6.0])

# 母分散（問題文で与えられている）
sigma1_sq = 1.2
sigma2_sq = 3.0

n1 = len(group1)
n2 = len(group2)

mean1 = group1.mean()
mean2 = group2.mean()

# -- z値の計算 ---
z = (mean1 - mean2) / np.sqrt(sigma1_sq/n1 + sigma2_sq/n2)

# -- p値（両側） ---
p_value = 2 * (1 - norm.cdf(abs(z)))

# -- 結果表示 ---
print("1組の平均:", mean1)
print("2組の平均:", mean2)
print("z値:", z)
print("p値:", p_value)

alpha = 0.05
if p_value < alpha:
    print("帰無仮説 H0 棄却 → 母平均は等しくない（差がある）")
else:
    print("帰無仮説 H0 採択 → 母平均は等しいと言える")
```

### ■ 問12　母平均の差の検定(対応のないデータ、分散未知だが等しい場合)

```python
import numpy as np
from scipy.stats import ttest_ind

# データ
group1 = np.array([8.0, 5.4, 4.8, 5.8, 6.4, 5.6])
group2 = np.array([9.0, 7.0, 3.6, 6.4, 6.0])

# 等分散 t検定（両側検定）
t_stat, p_value = ttest_ind(group1, group2, equal_var=True)

print("t値:", t_stat)
print("p値:", p_value)

alpha = 0.05

if p_value < alpha:
    print("→ 帰無仮説を棄却：1組と2組の母平均は等しくない（差がある）")
else:
    print("→ 帰無仮説を棄却できない：1組と2組の母平均は等しいとみなせる（差は有意ではない）")
```

### ■ 問14　母平均の差の検定(対応のないデータ、分散未知で等しいかどうかわからない場合)

```python
import numpy as np
from scipy.stats import ttest_ind

# データ
class1 = np.array([8.0, 5.4, 4.8, 5.8, 6.4, 5.6])
class2 = np.array([9.0, 7.0, 3.6, 6.4, 6.0])

# Welch の t 検定（等分散を仮定しない）
t_stat, p_value = ttest_ind(class1, class2, equal_var=False)

print("t値:", t_stat)
print("p値:", p_value)

# 有意水準5%で判断
alpha = 0.05
if p_value < alpha:
    print("帰無仮説を棄却 → 1組と2組の母平均は等しくないといえる")
else:
    print("帰無仮説を棄却できない → 平均に有意な差はない")
```

### ■ 問16　母分散の比の検定

```python
import numpy as np
from scipy.stats import f

# データ
A = np.array([10, 9, 8, 9, 10])
B = np.array([10, 10, 9, 9, 9, 10])

# 標本分散
var_A = np.var(A, ddof=1)
var_B = np.var(B, ddof=1)

# 標本数
nA = len(A)
nB = len(B)

# F値（大きい方の分散を分子にする）
if var_A >= var_B:
    F = var_A / var_B
    df1 = nA - 1
    df2 = nB - 1
else:
    F = var_B / var_A
    df1 = nB - 1
    df2 = nA - 1

# 有意水準
alpha = 0.05

# 臨界値（両側なので2.5%と97.5%点）
F_lower = f.ppf(alpha / 2, df1, df2)
F_upper = f.ppf(1 - alpha / 2, df1, df2)

# p値
p_value = 2 * min(1 - f.cdf(F, df1, df2), f.cdf(F, df1, df2))

# 結果表示
print("装置Aの標本分散:", var_A)
print("装置Bの標本分散:", var_B)
print(f"F値: {F:.4f}")
print(f"自由度 df1={df1}, df2={df2}")

print("\n--- 棄却域（両側）---")
print(f"F <= {F_lower:.4f}  または  F >= {F_upper:.4f}")

print("\n--- 判定 ---")
if F <= F_lower or F >= F_upper:
    print("棄却域に入るため、H0（母分散は等しい）を棄却。")
else:
    print("棄却域に入らないため、H0を棄却できない。")

print(f"\np値: {p_value:.4f}")
```
